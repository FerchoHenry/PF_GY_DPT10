# Analisis de cantidades de reviews

```python
print(f'El rango de reviews va desde:', dfc['review_count'].min(), 'hasta:', dfc['review_count'].max())
```

```python
qty = dfc[dfc['review_count'] < 7568]
qty
```

```python
plt.figure(figsize=(12,6))
plt.hist(dfc['review_count'], bins=2 , color = "lightgreen") 
plt.xlabel('Cuenta de reviews')
plt.ylabel('Frequencia')
plt.title('Distribucion de cantidades de reviews')
plt.show()
```

### En un primer analisis se puede observar que la mayoria de cantidad de reviews estan por abajo del los 4000

```python
#cantidad de restaurante con reviews entre 500 y el maximo de 7568
dfc['review_count'].where((dfc['review_count'] > 500) & (dfc['review_count'] <= 7568)).count()
```

Solo un porcentaje muy chico de los restaurantes cuentan con mas de 500 reviews (1348 de 150341 -> 0.89% )

```python
#limito la cantidad de reviews hasta 100 para un mejor analisis
cant = dfc['review_count'].where(dfc['review_count'] < 100) 
```

```python
plt.figure(figsize=(12,6))
plt.hist(cant, bins=10 ,color = "lightgreen") 
plt.xlabel('Cuenta de reviews')
plt.ylabel('Frequencia')
plt.title('Distribucion de cantidades de reviews')
plt.show()
```

Del grafico se puede observar que la mayoria de los negocios cuenta con menos de 20 reviews, con la media en 10 reviews

```python
plt.figure(figsize=(12,6))
plt.boxplot(dfc['review_count'].where(dfc['review_count'] < 90), showmeans=True) 
plt.xlabel('Cuenta de reviews')
plt.ylabel('Frequencia')
plt.title('Distribucion de cantidades de reviews')
plt.show()
```

comprobacion de media aproximada en 10 reviews
